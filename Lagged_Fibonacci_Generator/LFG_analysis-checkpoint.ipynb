# Implementation of Lagged Fibonacci Generator and plotting normal distribution

```python
# LFG random number generating

pop = 2000000 #population count

# random number based on mili seconds as seed
millis = int(round(time.time() * 10**45))
seed = list(map(int, str(millis)))

j = 24
k = 55
s = seed

mod = 1000 #cut's begining character of the number higher than 1000, 1258 -> 258

# implementing LFG algorithm

lfg_num =[]

for n in range(int(pop)):
        op = (s[j-1] + s[k-1]) % mod # wynik
        lfg_num.append(op)
        s.append(op)
        s.pop(0)

# algorithm for preparing normal distribution for generated numbers out of LFG list (lfg_mum)

sample = 50 # sample size
lfg_num_norm = []
x2=0

while x2 < len(lfg_num):
    avg = round((sum(random.sample(lfg_num,sample)))/sample)
    lfg_num_norm.append( avg )
    x2+=1

# preparing graphs
plt.rcParams['axes.grid'] = True
f = plt.figure(figsize=(15,5))
ax = f.add_subplot(121)
ax2 = f.add_subplot(122)

# distribution of generated numbers in LFG list (lfg_mum)
ax.hist(lfg_num,bins=50)
ax.title.set_text("Population: " + str(pop))
ax.set_ylabel("Occurrences")
ax.set_xlabel("Range 0-"+str(mod-1))

# normal distribution
ax2.hist(lfg_num_norm,bins=50)
ax2.title.set_text(str(sample) + " sample distribution")
ax2.set_ylabel("Occurrences")
ax2.set_xlabel("Range")

print('seed: ' + str(seed), 'seed len: ' + str(len(seed)), 
      #'\nseed7: ' + str(seed7), 'seed7 len: ' + str(len(seed7)), 
      '\nlen lfg_num: '+str(len(lfg_num)), 
      '\nsJ: ' +str(s[j-1]), 'sK: '+ str(s[k-1]))
```

# Genereting random numbers from Python's build-in funcion and plotting normal distribution

```python
# random number generating
rand_num = []

for i in range(0,int(pop)):
    pyth = round(random.uniform(0,999))
    rand_num.append(pyth)

# choosing samples form the number list to plot a distribution 
sample = 50
rand_num_norm = []
x3=0

while x3 < len(rand_num):
    avg = round((sum(random.sample(rand_num,sample )))/sample)
    rand_num_norm.append(avg)
    x3+=1

# graphs as previous
f = plt.figure(figsize=(15,5))
ax3 = f.add_subplot(121)
ax4 = f.add_subplot(122)

ax3.hist(rand_num,bins=50)
ax3.title.set_text("Population: " + str(pop))
ax3.set_ylabel("Occurrences")
ax3.set_xlabel("Range 0-"+str(mod-1))

ax4.hist(rand_num_norm,bins=50)
ax4.title.set_text(str(sample) + " sample distribution")
ax4.set_ylabel("Occurrences")
ax4.set_xlabel("Range")
```

### T-student test for independent groups

```python
from scipy.stats import ttest_ind
from scipy import stats
from scipy.stats import t
```

```python
# source https://machinelearningmastery.com/how-to-code-the-students-t-test-from-scratch-in-python/

def independent_ttest(lfg_num_norm, rand_num_norm, alpha):
    # calculate means
    mean1, mean2 = mean(lfg_num_norm), mean(rand_num_norm)
    # calculate standard errors
    se1, se2 = stats.sem(lfg_num_norm), stats.sem(rand_num_norm)
    # standard error on the difference between the samples
    sed = sqrt(se1**2.0 + se2**2.0)
    # calculate the t statistic
    t_stat = (mean1 - mean2) / sed
    # degrees of freedom
    df = len(lfg_num_norm) + len(rand_num_norm) - 2
    # calculate the critical value
    cv = t.ppf(1.0 - alpha, df)
    # calculate the p-value
    p = (1.0 - t.cdf(abs(t_stat), df)) * 2.0
    # return everything
    return t_stat, df, cv, p

alpha = 0.05
t_stat, df, cv, p = independent_ttest(lfg_num_norm, rand_num_norm, alpha)
print('t=%.3f, df=%d, cv=%.3f, p=%.3f' % (t_stat, df, cv, p))
# interpret via critical value
if abs(t_stat) <= cv:
    print('Accept null hypothesis that the means are equal. T_stat <= cv')
else:
    print('Reject the null hypothesis that the means are equal. T_stat > cv')
# interpret via p-value
if p > alpha:
    print('Accept null hypothesis that the means are equal. p > alpha')
else:
    print('Reject the null hypothesis that the means are equal. p <= alpha')
```

### Chisquare

```python
#http://specminor.org/2017/01/08/performing-chi-squared-gof-python.html
from collections import Counter
from scipy.stats import chisquare

lfg_num_np = np.array(lfg_num) # LFG numbers list
rand_num_np = np.array(rand_num) # Python numbers generated list

# counting occurences in lists (additional)
lfg_num_count = Counter(lfg_num_np)
rand_num_count = Counter(rand_num_np)

# counting occurences in lists and creating dictionaries 
unique, counts = np.unique(lfg_num_np, return_counts=True)
lfg_num_dict = dict(zip(unique, counts))

unique, counts = np.unique(rand_num_np, return_counts=True)
rand_num_dict = dict(zip(unique, counts))

# listing values
lfg_num_dict_vlist = list(lfg_num_dict.values())
rand_num_dict_vlist = list(rand_num_dict.values())

# Chi^2 statistics
chisquare(lfg_num_dict_vlist, f_exp=rand_num_dict_vlist)
```

```python
#building step by step Chisquare test

# peparing dictionaries and trapose
lfg_num_df = pd.DataFrame.from_dict([lfg_num_dict])
lfg_num_dfT = lfg_num_df.T

rand_num_df = pd.DataFrame.from_dict([rand_num_dict])
rand_num_dfT = rand_num_df.T

#column naming 
lfg_num_dfT = lfg_num_dfT.rename(columns={0:'count'})
rand_num_dfT = rand_num_dfT.rename(columns={0:'count'})

#gathering data for putting into Chisquare formula
observed = lfg_num_dfT
rand_num_ratio = rand_num_dfT/len(rand_num_dfT)
expected = rand_num_ratio * len(lfg_num_dfT)
chi_squared_stat = (((observed-expected)**2)/expected).sum()

# counting p-value
crit = stats.chi2.ppf(q = 0.95, # critical value for 95% confidence*
                      df = 999)   # Df = number of variable categories - 1

p_value = 1 - stats.chi2.cdf(x=chi_squared_stat,  # Finding the p-value
                             df=999)

print('Chi^2 stat: ' + str(chi_squared_stat))
print("Critical value :" + str(crit))
print("P value" + str(p_value))
stats.chisquare(f_obs= observed,   # Array of observed counts
                f_exp= expected)   # Array of expected counts
```
